# Fuzzy C-Mean Clustering

  ### Fuzzy C-Means adalah salah satu teknik clustering yang berdasarkan prinsip fuzzy logic, di mana setiap data tidak hanya menjadi bagian dari satu kelompok secara tegas, tetapi bisa menjadi anggota dari beberapa kluster sekaligus dengan tingkat keanggotaan tertentu. Ini membedakannya dari metode seperti K-Means yang hanya membolehkan keanggotaan tunggal.
---
## **Konsep Dasar Fuzzy C-Means Clustering**
  Fuzzy C-Means (FCM) adalah salah satu algoritma clustering atau pengelompokan data yang menggunakan pendekatan logika fuzzy, di mana satu data dapat termasuk dalam lebih dari satu kluster pada saat yang sama, dengan tingkat keanggotaan tertentu (nilai antara 0 dan 1).

  Beberapa komponen penting dalam algoritma FCM meliputi:
  - Titik Data (Data Point): Sekumpulan objek yang akan diklasifikasikan ke dalam beberapa kelompok.

  - Titik Pusat (Centroid): Titik yang berfungsi sebagai pusat dari masing-masing kelompok dalam proses pengelompokan.

  - Matriks Keanggotaan Fuzzy (U): Matriks yang memuat derajat keanggotaan setiap titik data terhadap masing-masing kelompok.

  - Parameter Fuzziness (m): Parameter yang menentukan tingkat ketidakpastian dalam keanggotaan suatu data terhadap kelompok tertentu. Umumnya bernilai 2; semakin besar nilai m, maka derajat keanggotaan data terhadap suatu kelompok menjadi semakin tidak tegas.

  
  **Perbedaan dengan K-Means**

  Pada metode klasterisasi seperti K-Means, setiap data hanya bisa menjadi anggota satu kluster secara pasti (hard clustering). Sedangkan dalam FCM, keanggotaan data terhadap kluster bersifat relatif (soft clustering) — artinya satu data bisa “tertarik” ke beberapa kluster, tetapi dengan bobot atau derajat keanggotaan yang berbeda-beda.

  ---

## **Algoritma Fuzzy C-Means Clustering**
 Fuzzy C-Means adalah algoritma *soft clustering* yang bekerja dengan cara memperbarui **pusat kluster** dan **nilai keanggotaan fuzzy** secara berulang, hingga hasilnya stabil. Setiap data dapat menjadi anggota lebih dari satu kluster, dengan nilai keanggotaan yang menunjukkan kedekatannya ke masing-masing pusat kluster.

---

### **Langkah-langkah Algoritma FCM**

#### **1. Inisialisasi**

* Tentukan jumlah kluster $c$
* Pilih nilai parameter *fuzziness* $m$ (biasanya $m = 2$)
* Buat matriks keanggotaan awal $U_{ij}$ secara acak, dengan syarat:

     $$
     \sum_{j=1}^c U_{ij} = 1, \quad \forall i
     $$


---

#### **2. Hitung Pusat Kluster**

$$
   v_j = \frac{\sum_{i=1}^n (U_{ij})^m \cdot x_i}{\sum_{i=1}^n (U_{ij})^m}
   $$

   Di mana:

   * $v_j$ adalah pusat cluster ke-j
   * $x_i$ adalah data ke-i
   * $U_{ij}$ adalah nilai keanggotaan data ke-i terhadap cluster ke-j

---

#### **3. Update derajat keanggotaan**

Untuk setiap data $x_i$ dan kluster $j$, perbarui derajat keanggotaan:

$$
u_{ij} = \frac{1}{\sum_{k=1}^{c} \left( \frac{\|x_i - c_j\|}{\|x_i - c_k\|} \right)^{\frac{2}{m - 1}}}
$$

Di mana:

   * $\|x_i - v_j\|$ adalah jarak antara data ke-i dengan pusat cluster ke-j
---

#### **4. Cek Konvergensi**

* Hitung perubahan antara matriks keanggotaan sebelumnya dan yang baru.
* Jika selisihnya lebih kecil dari nilai toleransi $\epsilon$, hentikan iterasi.
* Jika belum, ulangi langkah 2 dan 3.

---

#### **5. Penentuan Hasil**

* Setiap data dikaitkan dengan kluster yang memiliki **nilai keanggotaan tertinggi**.
* Hasil akhirnya adalah:

  * Pusat kluster akhir
  * Matriks keanggotaan akhir
  * Label kluster dominan untuk tiap data (jika diperlukan)

---

## **Contoh Manual Perhitungan Fuzzy C-Means Clustering**

Misalkan data dan parameter:

### **Data:**

$$
X = \{1, 2, 5, 6\}
$$

### **Jumlah Cluster (c):** 2
### **Tingkat Fuzziness (m):** 2
### **Kriteria Konvergensi:** perbedaan pusat cluster < 0.01

---

## **Langkah 1: Inisialisasi Matriks Keanggotaan $U$**

Ukuran matriks: $2 \times 4$

Misalkan mulai dengan matriks:

$$
U^{(0)} =
\begin{bmatrix}
0.3 & 0.2 & 0.4 & 0.1 \\
0.7 & 0.8 & 0.6 & 0.9 \\
\end{bmatrix}
$$

Sudah memenuhi syarat jumlah keanggotaan tiap data = 1.

---

## **Langkah 2: Hitung Pusat Cluster (Centroid)**

Rumus:

$$
v_j = \frac{\sum_{i=1}^n u_{ji}^m \cdot x_i}{\sum_{i=1}^n u_{ji}^m}
$$

Hitung untuk cluster 1:

$$
\begin{aligned}
v_1 &= \frac{0.3^2 \times 1 + 0.2^2 \times 2 + 0.4^2 \times 5 + 0.1^2 \times 6}{0.3^2 + 0.2^2 + 0.4^2 + 0.1^2} \\
&= \frac{0.09 \times 1 + 0.04 \times 2 + 0.16 \times 5 + 0.01 \times 6}{0.09 + 0.04 + 0.16 + 0.01} \\
&= \frac{0.09 + 0.08 + 0.8 + 0.06}{0.3} \\
&= \frac{1.03}{0.3} \\
&\approx 3.433
\end{aligned}
$$

Hitung untuk cluster 2:

$$
\begin{aligned}
v_2 &= \frac{0.7^2 \times 1 + 0.8^2 \times 2 + 0.6^2 \times 5 + 0.9^2 \times 6}{0.7^2 + 0.8^2 + 0.6^2 + 0.9^2} \\
&= \frac{0.49 \times 1 + 0.64 \times 2 + 0.36 \times 5 + 0.81 \times 6}{0.49 + 0.64 + 0.36 + 0.81} \\
&= \frac{0.49 + 1.28 + 1.8 + 4.86}{2.3} \\
&= \frac{8.43}{2.3} \\
&\approx 3.667
\end{aligned}
$$

---

## **Langkah 3: Update Matriks Keanggotaan $U$**

Rumus:

$$
u_{ji} = \frac{1}{\sum_{k=1}^c \left(\frac{d(x_i, v_j)}{d(x_i, v_k)}\right)^{\frac{2}{m-1}}}
$$

Hitung jarak untuk tiap data dan centroid:

| Data $x_i$ | $d(x_i, v_1)$ | $d(x_i, v_2)$ |          |   |           |          |
| ---------- | ------------- | ------------- | -------- | - | --------- | -------- |
| 1          | (             | 1 - 3.433     | = 2.433) | ( | 1 - 3.667 | = 2.667) |
| 2          | (             | 2 - 3.433     | = 1.433) | ( | 2 - 3.667 | = 1.667) |
| 5          | (             | 5 - 3.433     | = 1.567) | ( | 5 - 3.667 | = 1.333) |
| 6          | (             | 6 - 3.433     | = 2.567) | ( | 6 - 3.667 | = 2.333) |

Contoh update untuk data $x_1 = 1$:

$$
u_{1,1} = \frac{1}{\left(\frac{2.433}{2.433}\right)^2 + \left(\frac{2.433}{2.667}\right)^2}
= \frac{1}{1 + (0.912)^2} = \frac{1}{1 + 0.831} = \frac{1}{1.831} \approx 0.546
$$

$$
u_{2,1} = 1 - u_{1,1} = 0.454
$$

Ulangi proses ini untuk seluruh data:

| Data | $u_{1,i}$                                                                             | $u_{2,i}$ |
| ---- | ------------------------------------------------------------------------------------- | --------- |
| 1    | 0.546                                                                                 | 0.454     |
| 2    | $\frac{1}{1 + (1.433/1.667)^2} = \frac{1}{1 + 0.739^2} = \frac{1}{1 + 0.546} = 0.647$ | 0.353     |
| 5    | $\frac{1}{1 + (1.567/1.333)^2} = \frac{1}{1 + 1.175^2} = \frac{1}{1 + 1.381} = 0.420$ | 0.580     |
| 6    | $\frac{1}{1 + (2.567/2.333)^2} = \frac{1}{1 + 1.101^2} = \frac{1}{1 + 1.213} = 0.452$ | 0.548     |

---

## **Langkah 4: Cek Konvergensi**

Hitung perubahan centroid dari iterasi sebelumnya (kita hanya ada 1 iterasi):

* $v_1^{(0)} =$ (tidak ada, inisialisasi dari $U^{(0)}$)
* $v_1^{(1)} = 3.433$
* $v_2^{(1)} = 3.667$

Karena belum ada iterasi sebelumnya, lanjutkan iterasi 2 dengan $U^{(1)}$:

$$
U^{(1)} =
\begin{bmatrix}

0.454 & 0.353 & 0.580 & 0.548 \\
\end{bmatrix}
$$

## **Langkah 2 Iterasi 2: Hitung Centroid Baru**

$$
v_1^{(2)} = \frac{(0.546)^2 \cdot 1 + (0.647)^2 \cdot 2 + (0.420)^2 \cdot 5 + (0.452)^2 \cdot 6}{(0.546)^2 + (0.647)^2 + (0.420)^2 + (0.452)^2}
$$

$$
= \frac{0.298 \times 1 + 0.419 \times 2 + 0.176 \times 5 + 0.204 \times 6}{0.298 + 0.419 + 0.176 + 0.204}
= \frac{0.298 + 0.838 + 0.881 + 1.224}{1.097}
= \frac{3.241}{1.097}
\approx 2.955
$$

$$
v_2^{(2)} = \frac{(0.454)^2 \cdot 1 + (0.353)^2 \cdot 2 + (0.580)^2 \cdot 5 + (0.548)^2 \cdot 6}{(0.454)^2 + (0.353)^2 + (0.580)^2 + (0.548)^2}
$$

$$
= \frac{0.206 \times 1 + 0.125 \times 2 + 0.336 \times 5 + 0.300 \times 6}{0.206 + 0.125 + 0.336 + 0.300}

## **Langkah 3 Iterasi 2: Update Matriks Keanggotaan $U$**

Hitung jarak baru:

| Data | $d(x_i, v_1^{(2)})$ | $d(x_i, v_2^{(2)})$ |          |   |           |          |
| ---- | ------------------- | ------------------- | -------- | - | --------- | -------- |

Contoh update untuk data 1:

$$
u_{1,1} = \frac{1}{1 + \left(\frac{1.955}{3.071}\right)^2} = \frac{1}{1 + (0.637)^2} = \frac{1}{1 + 0.406} = 0.711
$$

$$
u_{2,1} = 1 - 0.711 = 0.289

Ulangi seluruh data:

* Fungsi objektif menurun dari iterasi awal ke iterasi berikutnya (harus dihitung ulang secara teliti).

## **Implementasi Python**

```python
import numpy as np
import pandas as pd

# Data dan parameter
X = np.array([1, 2, 5, 6])  # Data satu dimensi
c = 2                      # Jumlah cluster
m = 2.0                    # Tingkat fuzziness
epsilon = 0.01             # Ambang batas konvergensi
max_iter = 100             # Maksimum iterasi

# Inisialisasi keanggotaan U secara manual (2 x 4)
U = np.array([
    [0.3, 0.2, 0.4, 0.1],
    [0.7, 0.8, 0.6, 0.9]
])

# Fungsi untuk menghitung pusat cluster
def calculate_centroids(U, X, m):
    um = U ** m
    return (um @ X) / um.sum(axis=1)

# Fungsi untuk update matriks keanggotaan U
def update_membership(X, V, m):
    n = len(X)
    c = len(V)
    U_new = np.zeros((c, n))
    for i in range(n):
        for j in range(c):
            denom_sum = 0.0
            for k in range(c):
                dist_j = abs(X[i] - V[j])
                dist_k = abs(X[i] - V[k])
                if dist_k == 0:
                    dist_k = 1e-10  # menghindari pembagian nol
                dist_ratio = dist_j / dist_k
                denom_sum += dist_ratio ** (2 / (m - 1))
            U_new[j, i] = 1.0 / denom_sum
    return U_new

# Fungsi objektif
def objective_function(U, V, X, m):
    obj = 0.0
    for j in range(c):
        for i in range(len(X)):
            obj += (U[j, i] ** m) * (abs(X[i] - V[j]) ** 2)
    return obj

# Iterasi sampai konvergen
history = []
for iteration in range(max_iter):
    V = calculate_centroids(U, X, m)
    obj = objective_function(U, V, X, m)
    history.append((iteration+1, V.copy(), U.copy(), obj))

    U_new = update_membership(X, V, m)

    # Cek konvergensi berdasarkan perubahan centroid
    if np.max(np.abs(V - calculate_centroids(U_new, X, m))) < epsilon:
        break
    U = U_new

# Konversi hasil ke DataFrame agar mudah dibaca
results = []
for iter_num, centroids, membership, obj_val in history:
    result = {
        "Iterasi": iter_num,
        "Centroid 1": centroids[0],
        "Centroid 2": centroids[1],
        "Objektif": obj_val
    }
    for j in range(c):
        for i in range(len(X)):
            result[f"U[{j+1},{i+1}]"] = membership[j, i]
    results.append(result)

df_results = pd.DataFrame(results)
print(df_results.round(4))
```

## **Clustering menggunakan Fuzzy C-Means**

### Clustering dengan 2 cluster dan Visualisasi data

---
Dalam proses ini dilakukan clustering menggunakan metode Fuzzy C-Means dengan jumlah cluster sebanyak 2. Selain itu, juga dihitung pusat cluster dan fungsi objektif untuk setiap iterasi yang terjadi. Selanjutnya, hasil clustering divisualisasikan dalam bentuk plot dua dimensi menggunakan data Iris.

```python

# Transpose fitur agar sesuai dengan input skfuzzy (fitur x sample)
data = scaled_features.T

# Fuzzy C-Means clustering
n_clusters = 2
cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(
    data, c=n_clusters, m=2.0, error=0.005, maxiter=1000, init=None, seed=42
)

# Ambil label keras dari keanggotaan tertinggi
labels_2 = np.argmax(u, axis=0)

# Evaluasi clustering
silhouette_fuzzy = silhouette_score(scaled_features, labels_2)

# Cetak hasil evaluasi
print(f"[Fuzzy C-Means - 2 Clusters]")
print(f"Silhouette Score: {silhouette_fuzzy:.4f}")
print(f"Fuzzy Partition Coefficient (FPC): {fpc:.4f}")
print(f"Objective Function Value (Jm): {jm[-1]:.4f}")  # ambil nilai terakhir iterasi
print(f"Cluster Centers (in scaled feature space):\n{cntr}")

# Menampilkan nilai fungsi objektif di setiap iterasi
print("Objective Function Values per Iteration:")
for i, val in enumerate(jm):
    print(f"Iterasi {i+1:02d}: {val:.4f}")


# Visualisasi hasil clustering (dalam ruang PCA)
plt.figure(figsize=(6, 4))
plt.scatter(reduced_features[:, 0], reduced_features[:, 1], c=labels_2, cmap='viridis', s=50)
plt.title("Fuzzy C-Means Clustering (2 Clusters)")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.grid(True)
plt.show()
```

Kode ini menggunakan metode Fuzzy C-Means Clustering pada dataset Iris yang sudah melalui proses scaling, kemudian mengurangi dimensi data menjadi tiga komponen utama menggunakan PCA untuk tujuan visualisasi 3D, sekaligus melakukan evaluasi hasil klastering. Pertama, data numerik yang telah dinormalisasi diproyeksikan ke ruang berdimensi tiga dan disimpan dalam variabel `reduced_features_3d`. Data tersebut kemudian ditranspos agar sesuai dengan format input fungsi `cmeans` dari pustaka `skfuzzy`, yaitu fitur sebagai baris dan sampel sebagai kolom. Proses Fuzzy C-Means dijalankan dengan parameter dua klaster (n\_clusters = 2) dan tingkat keburaman m = 2.0. Fungsi ini menghasilkan pusat cluster (`cntr`), matriks keanggotaan (`u`), serta metrik lain seperti nilai fungsi objektif tiap iterasi (`jm`) dan Fuzzy Partition Coefficient (`fpc`). Label klaster "keras" ditentukan berdasarkan nilai keanggotaan tertinggi untuk masing-masing sampel. Untuk mengukur kualitas klastering, digunakan Silhouette Score, dan hasil evaluasi, termasuk nilai FPC dan nilai fungsi objektif terakhir, ditampilkan di konsol. Terakhir, hasil klaster divisualisasikan dalam bentuk scatter plot 3D berdasarkan hasil reduksi PCA dengan pewarnaan sesuai label klaster yang diperoleh.

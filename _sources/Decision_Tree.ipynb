## **2. Entropi Dataset Awal (Parent Node)**

### Total Data:

* Kelas A = 6
* Kelas B = 9

$$
\begin{gathered}
p_A = \frac{6}{15}, \quad p_B = \frac{9}{15} \\
\text{Entropy}(S) = -\left( \frac{6}{15} \log_2 \frac{6}{15} + \frac{9}{15} \log_2 \frac{9}{15} \right) \\
= -\left( 0.4 \cdot \log_2 0.4 + 0.6 \cdot \log_2 0.6 \right) \\
= -\left( 0.4 \cdot (-1.322) + 0.6 \cdot (-0.737) \right) \\
= -( -0.5288 - 0.4422 ) = 0.5288 + 0.4422 = \boxed{0.971}
\end{gathered}
$$

## **3. Split Berdasarkan Fitur: "Panjang Tubuh (cm)"**

Misal kita pilih split sederhana: **Panjang ≤ 235**

### **Entropy (Left Node: Panjang ≤ 235)**

$$
p_A = \frac{2}{11} \approx 0.182,\quad p_B = \frac{9}{11} \approx 0.818
$$

$$
\text{Entropy}(S_{\text{left}}) = -\left( p_A \cdot \log_2 p_A + p_B \cdot \log_2 p_B \right)
$$

Gunakan pembulatan logaritma (3 digit):

* $\log_2 0.182 \approx -2.459$
* $\log_2 0.818 \approx -0.289$

$$
= -\left( 0.182 \cdot (-2.459) + 0.818 \cdot (-0.289) \right)
$$

$$
= -\left( -0.4476 - 0.2363 \right)
= 0.4476 + 0.2363 = \boxed{0.684}
$$

---
### **Hasil Akhir:**

$$
\boxed{\text{Entropy (Left)} = 0.684}
$$

### **Grup Kanan (Panjang > 235)**

#### **Entropy(Right):**

$$
\operatorname{Entropy}(\text{Right}) = -\left(1 \cdot \log_2 1 + 0 \cdot \log_2 0\right) = -0 = \boxed{0}
$$

---

#### **Information Gain (Panjang > 235)**

$$
\begin{aligned}
IG &= \operatorname{Entropy}(S) - \left(\frac{11}{15} \cdot \operatorname{Entropy}(\text{Left}) + \frac{4}{15} \cdot \operatorname{Entropy}(\text{Right})\right) \\
&= 0.971 - \left(\frac{11}{15} \cdot 0.684 + \frac{4}{15} \cdot 0\right) \\
&= 0.971 - (0.5027 + 0) = \boxed{0.469}
\end{aligned}
$$

#### **Hitung Entropy (Grup Left)**

$$
\begin{align*}
p_A &= \frac{6}{13} \approx 0.462, \\
p_B &= \frac{7}{13} \approx 0.538 \\
\\
\text{Entropy (Left)} &= -\left(p_A \cdot \log_2 p_A + p_B \cdot \log_2 p_B\right) \\
&= -\left(0.462 \cdot \log_2 0.462 + 0.538 \cdot \log_2 0.538\right) \\
&= -\left(0.462 \cdot -1.112 + 0.538 \cdot -0.891\right) \\
&= 0.513 + 0.479 \\
&= \mathbf{0.992}
\end{align*}
$$

**Grup Kanan (Berat > 215):**

| Berat | Kelas |
| ----- | ----- |
| 216   | B     |
| 218   | B     |


* **Total = 2**
* **A = 0**
* **B = 2**


#### **Entropy Grup Kanan (> 215)**

$$
\text{Entropy(Right)} = -\left(\frac{0}{2} \log_2 \frac{0}{2} + \frac{2}{2} \log_2 \frac{2}{2}\right) = -\left(0 + 1 \cdot \log_2 1\right) = 0
$$

$$
\boxed{\text{Entropy(Right)} = 0}
$$

---

#### **Langkah 3: Information Gain (IG)**

$$
IG = \text{Entropy}(S) - \left(\frac{13}{15} \cdot \text{Entropy(Left)} + \frac{2}{15} \cdot \text{Entropy(Right)}\right)
$$

Diketahui:

* $\text{Entropy}(S) = 0.971$ (dari awal)
* $\text{Entropy(Left)} = 0.994$
* $\text{Entropy(Right)} = 0$

$$
IG = 0.971 - \left(\frac{13}{15} \cdot 0.994 + \frac{2}{15} \cdot 0\right) = 0.971 - 0.861 = \boxed{0.110}
$$

---

## **Visualisasi Decision Tree**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.preprocessing import LabelEncoder

# === Data sesuai dataset Anda ===
data = {
    "Panjang (cm)": [225, 235, 240, 228, 237, 220, 215, 212, 218, 217, 213, 243, 230, 214, 241],
    "Berat (kg)": [205, 210, 208, 215, 212, 218, 214, 209, 207, 206, 203, 211, 216, 202, 209],
    "Kelas": ['A', 'A', 'A', 'B', 'A', 'B', 'B', 'B', 'B', 'B', 'B', 'A', 'B', 'B', 'A']
}

df = pd.DataFrame(data)

# === Encode label A/B menjadi 0/1 ===
le = LabelEncoder()
df['Label'] = le.fit_transform(df['Kelas'])  # A=0, B=1

# === Fitur dan Target ===
X = df[["Panjang (cm)", "Berat (kg)"]]
y = df["Label"]

# === Train Decision Tree ===
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3, random_state=42)
clf.fit(X, y)

# === Tampilkan aturan pohon (rules) ===
print("\n=== Aturan Pohon Keputusan ===")
tree_rules = export_text(clf, feature_names=list(X.columns))
print(tree_rules)

# === Visualisasi Pohon Keputusan ===
plt.figure(figsize=(12, 6))
plot_tree(clf,
          feature_names=["Panjang (cm)", "Berat (kg)"],
          class_names=le.classes_,  # ['A', 'B']
          filled=True,
          rounded=True,
          fontsize=10)
plt.title("Visualisasi Pohon Keputusan")
plt.show()
```

## **Visualisasi Splitting Setelah ada 2 Data Baru**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.preprocessing import LabelEncoder

# === Data ===
data = {
    "Panjang (cm)": [225, 235, 240, 228, 237, 220, 215, 212, 218, 217, 213, 243, 230, 214, 241],
    "Berat (kg)": [205, 210, 208, 215, 212, 218, 214, 209, 207, 206, 203, 211, 216, 202, 209],
    "Kelas": ['A', 'A', 'A', 'B', 'A', 'B', 'B', 'B', 'B', 'B', 'B', 'A', 'B', 'B', 'A']
}

df = pd.DataFrame(data)

# === Encode label A/B menjadi 0/1 ===
le = LabelEncoder()
df['Label'] = le.fit_transform(df['Kelas'])  # A=0, B=1

# === Fitur dan Target ===
X = df[["Panjang (cm)", "Berat (kg)"]]
y = df["Label"]

# === Latih Decision Tree ===
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3, random_state=42)
clf.fit(X, y)

# === Prediksi 2 Data Baru ===
new_data = pd.DataFrame({
    "Panjang (cm)": [236, 216],
    "Berat (kg)": [210, 204]
})
preds = clf.predict(new_data)
predicted_labels = le.inverse_transform(preds)

# === Visualisasi Decision Boundary ===
x_min, x_max = X["Panjang (cm)"].min() - 5, X["Panjang (cm)"].max() + 5
y_min, y_max = X["Berat (kg)"].min() - 5, X["Berat (kg)"].max() + 5
xx, yy = np.meshgrid(np.arange(x_min, x_max, 1),
                     np.arange(y_min, y_max, 1))

Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
plt.scatter(X["Panjang (cm)"], X["Berat (kg)"], c=y, cmap=plt.cm.coolwarm, edgecolor="k", s=60, label="Data Latih")
plt.scatter(new_data["Panjang (cm)"], new_data["Berat (kg)"], c="yellow", edgecolor="black", s=100, label="Data Baru")
for i, label in enumerate(predicted_labels):
    plt.text(new_data.iloc[i, 0]+0.5, new_data.iloc[i, 1], label, fontsize=10, weight='bold')

plt.xlabel("Panjang (cm)")
plt.ylabel("Berat (kg)")
plt.title("Decision Boundary Berdasarkan Dataset Kamu")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Tampilkan hasil prediksi
print("Prediksi Data Baru:", predicted_labels)

# Tampilkan aturan pohon keputusan
print("\n=== Aturan Pohon Keputusan ===")
print(export_text(clf, feature_names=X.columns.tolist()))
```
